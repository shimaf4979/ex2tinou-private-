```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
stones = list(range(2, 21))
visited = [
    3,
    6,
    12,
    22,
    23,
    46,
    92,
    162,
    163,
    326,
    652,
    1142,
    1143,
    2286,
    4572,
    8002,
    8003,
    16006,
    32012,
]
cuts = [
    1,
    2,
    4,
    7,
    8,
    16,
    32,
    56,
    57,
    114,
    228,
    399,
    400,
    800,
    1600,
    2800,
    2801,
    5602,
    11204,
]

# Graph style settings
plt.style.use("seaborn")
plt.figure(figsize=(15, 10))

# Subplot 1: Number of visited nodes
plt.subplot(2, 1, 1)
plt.plot(stones, visited, marker="o", linewidth=2, markersize=8)
plt.grid(True)
plt.title(
    "Relationship between Number of Stones and Visited Nodes", fontsize=14, pad=15
)
plt.xlabel("Number of Stones", fontsize=12)
plt.ylabel("Number of Visited Nodes", fontsize=12)
plt.yscale("log")
plt.tick_params(axis="both", which="major", labelsize=10)

# Subplot 2: Number of pruning occurrences
plt.subplot(2, 1, 2)
plt.plot(stones, cuts, marker="o", color="red", linewidth=2, markersize=8)
plt.grid(True)
plt.title(
    "Relationship between Number of Stones and Pruning Occurrences", fontsize=14, pad=15
)
plt.xlabel("Number of Stones", fontsize=12)
plt.ylabel("Number of Pruning Occurrences", fontsize=12)
plt.yscale("log")
plt.tick_params(axis="both", which="major", labelsize=10)

plt.tight_layout()
plt.show()
```
